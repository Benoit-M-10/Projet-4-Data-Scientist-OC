# <a name="C2">Création du jeu de données final utilisé pour la modélisation

Pour tester les modèles de clustering, nous avons choisi d'utiliser pour chaque client les variables suivantes :  
- Nombre de commandes par client
- La récence de la dernière commande par client
- La moyenne du montant payé par commande par client
- La moyenne des échéances de paiement par client
- La moyenne des notes de satisfaction par client
- La moyenne des % de ‘vouchers’ utilisés pour payer une commande par client

Nous allons donc créer ce dataframe récapitulatif à partir des dataframes que l'on a importés.

```python
# On part du dataframe sur les paiements.

df_order_payments
```

```python
df_order_payments.isna().sum()
```

```python
df_order_payments.describe(include="all")
```

```python
df_order_payments["payment_type"].unique()
```

```python
# On analyse les 'payment_type' qui sont en 'not_defined'.

df_order_payments[df_order_payments["payment_type"] == "not_defined"]
```

```python
# On supprime les 'payment_type' qui sont en 'not_defined'.

df_order_payments = df_order_payments[
    df_order_payments["payment_type"] != "not_defined"
]
```

```python
# On analyse les 'payment_installments' qui sont à 0.

df_order_payments[df_order_payments["payment_installments"] == 0]
```

```python
# On les remplace par 1.

df_order_payments["payment_installments"] = df_order_payments[
    "payment_installments"
].replace(0, 1)
```

```python
# On analyse les 'payment_value' qui sont à 0.

df_order_payments[df_order_payments["payment_value"] == 0]
```

```python
# On peut les conserver car les 'payment_sequential' sont différents de 1 donc le montant de la commande totale sera
# bien supérieur à 0 pour ces commandes.
```

```python
# On crée un pivot table pour pouvoir ensuite calculer le montant total payé pour chaque commande en sommant
# le montant payé pour chaque moyen de paiement. Cela permettra également de déterminer le 'payment_installments' maximum
# pour chaque commande ainsi que le % du montant total payé en voucher.

df_order_payments_pivot = pd.pivot_table(
    df_order_payments,
    index=["order_id"],
    columns=["payment_type"],
    values=["payment_value", "payment_installments"],
    aggfunc={"payment_value": np.sum, "payment_installments": np.max},
    fill_value=0,
)
```

```python
df_order_payments_pivot
```

```python
df_order_payments_pivot.loc["ea9184ad433a404df1d72fa0a8764232", :]
```

```python
# On détermine le 'payment_installments' maximum pour chaque commande.

df_order_payments_pivot[
    ("payment_installments", "max_payment_installments")
] = df_order_payments_pivot.iloc[:, 0:4].max(axis=1)
```

```python
df_order_payments_pivot
```

```python
# On calcule le montant total payé pour chaque commande en sommant le montant payé pour chaque moyen de paiement.

df_order_payments_pivot[
    ("payment_value", "sum_payment_value")
] = df_order_payments_pivot.iloc[:, 4:8].sum(axis=1)
```

```python
df_order_payments_pivot
```

```python
# On calcule le % du montant total payé en voucher.

df_order_payments_pivot[("payment_value", "ratio_voucher_payment_value")] = (
    df_order_payments_pivot[("payment_value", "voucher")]
    / df_order_payments_pivot[("payment_value", "sum_payment_value")]
)
```

```python
df_order_payments_pivot
```

```python
df_order_payments_pivot.iloc[:, -3:].describe()
```

```python
# On ne garde que les colonnes calculées pour pouvoir ensuite fusionner le dataframe obtenu avec le dataframe des orders.

df_order_payments_pivot = df_order_payments_pivot.loc[
    :,
    [
        ("payment_installments", "max_payment_installments"),
        ("payment_value", "sum_payment_value"),
        ("payment_value", "ratio_voucher_payment_value"),
    ],
].reset_index()
```

```python
df_order_payments_pivot
```

```python
df_order_payments_pivot.columns = [
    "order_id",
    "max_payment_installments",
    "sum_payment_value",
    "ratio_voucher_payment_value",
]
```

```python
df_order_payments_pivot
```

```python
df_orders = pd.read_csv("olist_orders_dataset.csv", sep=",")
```

```python
df_orders
```

```python
# On fusionne 'df_orders' avec 'df_order_payments_pivot'.

df_orders = pd.merge(df_orders, df_order_payments_pivot, on="order_id", how="inner")
```

```python
df_orders
```

```python
# On fusionne 'df_orders' avec 'df_customers'.

df_orders = pd.merge(df_orders, df_customers, on="customer_id", how="inner")
```

```python
df_orders
```

```python
df_orders["order_purchase_timestamp"] = pd.to_datetime(
    df_orders["order_purchase_timestamp"]
)
```

```python
df_orders["order_purchase_timestamp"].max()
```

```python
# On calcule la récence de chaque commande en mois par rapport à la dernière commande disponible dans le jeu de données (= la
# commande la plus récente).

df_orders["order_purchase_recency_in_months"] = -(
    df_orders["order_purchase_timestamp"] - df_orders["order_purchase_timestamp"].max()
) / np.timedelta64(1, "M")
```

```python
df_orders
```

```python
df_orders.isna().sum()
```

```python
# On crée un pivot table pour obtenir pour chaque client :
#      - son 'installments' moyen
#      - son montant de commande moyen
#      - son % de voucher moyen utilisé par rapport au montant total moyen de commande
#      - son 'review score' moyen
#      - sa récence la plus récente (= en cas de plusieurs commandes on prend la récence minimum)
#      - son nombre moyen de commandes

df_orders_pivot = pd.pivot_table(
    df_orders,
    index=["customer_unique_id"],
    values=[
        "order_id",
        "max_payment_installments",
        "sum_payment_value",
        "ratio_voucher_payment_value",
        "review_score",
        "order_purchase_recency_in_months",
    ],
    aggfunc={
        "order_id": len,
        "max_payment_installments": np.mean,
        "sum_payment_value": np.mean,
        "ratio_voucher_payment_value": np.mean,
        "review_score": np.mean,
        "order_purchase_recency_in_months": np.min,
    },
)
```

```python
df_orders_pivot
```

```python
df_orders_pivot = df_orders_pivot.reset_index()
```

```python
df_orders_pivot.columns = [
    "customer_unique_id",
    "mean_payment_installments",
    "number_of_orders",
    "last_order_purchase_recency_in_months",
    "mean_ratio_voucher",
    "mean_review_score",
    "mean_order_value",
]
```
